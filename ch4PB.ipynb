```python
#312
p="Rama#fo9c"
alpha=0
digit=0
special=0
no=0
if len(p)>=8:
    for i in p:
        if i.isalpha():
            if i.isupper():
                alpha+=1
        elif i.isdigit():
            digit+=1
        elif i=='_' or i=='@' or i=='$':
            special+=1
        else:
            no+=1
    if alpha>0 and digit>0 and special>0 and no==0:
        print("Valid Password")
    else:
        print("Inavlid Password")
else:
    print("Invalid Password")
```
