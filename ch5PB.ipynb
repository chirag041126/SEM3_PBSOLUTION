```python
l=[4,6,6,4,2,2,4,8,5,8]
l.sort()
count=0
prev=l[0]
ans={}
for i in l:
    if prev==i:
        count+=1
    else:
        ans[prev]=[prev]*count
        count=1
        prev=i
else:
    ans[prev]=[prev]*count
ans=dict(sorted(ans.items(), key=lambda x:len(x[1]),reverse=True))
print(ans)
```

```python
def combination(digits, start, end):
    if start==end:
        print(*digits)
    else:
        for i in range(start,end+1):
            digits[start],digits[i]=digits[i],digits[start]
            combination(digits,start+1,end)
            digits[start],digits[i]=digits[i],digits[start]

def unique(l):
    s=set(l)
    print("unique element: ",len(s))

def pro(l):
    ans=1
    s=set(l)
    for i in s:
        ans*=i
    print("Product of unique elements: ",ans)

def sumU(l):
    s=set(l)
    ans=sum(s)
    print("Sum of unique elements: ",ans)

num=[1,2,3]
combination(num,0,len(num)-1)
unique([1,2,2])
pro([2,2,3])
sumU([1,2,2])
```

```python
def A(l):
    t=tuple(map(lambda x: str(x), l))
    return t

def B(l):
    l=list(map(lambda x: x*10, l))
    return l

def C(l):
    from functools import reduce
    ans=reduce((lambda x,y: x*y), l)
    return ans

def D(s, ch):
    count=0
    for i in s:
        if i==ch:
            count+=1
    return count

def E(s, ch):
    for i in range(0, len(s)):
        if s[i]==ch:
            return i
    return -1

print(A([1,2,3,4]))
print(B([1,2,3,4]))
print(C([1,2,3,4]))
print(D("hello",'l'))
print(E("helloe",'e'))
print(E("helloe",'z'))
```

```python
#446
ones=["zero","one","two","three","four","five","six","seven","eight","nine"]
teens=["ten","eleven","twelve","thirteen","fourteen","fifteen","sixteen","seventeen","eighteen","ninteen"]
tens=["","","twenty","thirty","forty","fifty","sixty","seventy","eighty","ninty"]
def word(n):
    if 0<=n<=9:
        return ones[n]
    elif 10<=n<=19:
        return teens[n-10]
    elif 20<=n<100:
        if (n%10==0):
            return tens[n//10]
        else:
            return tens[n//10]+"-"+ones[n%10]
    else:
        if n%100==0:
            return ones[n//100]+ " hundred"
        else:
            return ones[n//100]+" hundred and "+ word(n%100)

def checkPassword(p):
    alphaU=0
    alphaL=0
    digit=0
    special=0
    no=0
    num=0
    if len(p)>=8 and len(p)<=15:
        for i in p:
            if i.isupper():
                alphaU+=1
            elif i.islower():
                alphaL+=1
            elif i.isdigit():
                num=num*10 + int(i)
                digit+=1
            elif i=='_' or i=='@' or i=='$':
                special+=1
            else:
                no+=1
        if alphaU>0 and alphaL>0 and digit<4 and special>0 and no==0:
            print("Valid Password")
            print("num: ",num)
            print(word(num))
        else:
            print("Inavlid Password")
    else:
        print("Invalid Password")
checkPassword(p="R@m@9fatu9e$9")
```

```python
stack=[]
pair={')':'(', ']':'[','}':'{'}
s='{()()'
for ch in s:
    if ch in '({[':
        stack.append(ch)
    elif ch in ')}]':
        if len(stack)==0 or stack[-1]!=pair[ch]:
            print("Invalid")
            break
        else:
            stack.pop()
    else:
        continue
else:
    if len(stack)==0:
        print("Valid")
    else:
        print("Invalid") 
```
